## Step 8: Setup Environment Variables

You need the following from your Clerk Dashboard (https://dashboard.clerk.com):

### Backend Configuration:
- **CLERK_API_KEY** - Get from https://dashboard.clerk.com (API Keys section, Secret Key)
- **CLERK_JWT_PUB_KEY** - Get from https://dashboard.clerk.com (JWKS URL)

### Frontend Configuration:
- **CLERK_PUBLISHABLE_KEY** - Get from https://dashboard.clerk.com (API Keys section, Publishable Key)
  - ⚠️ **IMPORTANT:** This is required for the frontend to work! Without it, you'll see a blank page.
  - Look for the key starting with `pk_test_` or `pk_live_`

### Optional:
- **TAVILY_API_KEY** (optional) - Get from https://tavily.com

PostgreSQL and LLM (Ollama) are running locally (no external URLs needed).

```python
from getpass import getpass

print("=" * 60)
print("CLERK CONFIGURATION")
print("=" * 60)

CLERK_API_KEY = getpass('Enter your Clerk API key: ')
CLERK_JWT_PUB_KEY = getpass('Enter your Clerk JWKS URL: ')
TAVILY_API_KEY = getpass('Enter your Tavily API key (optional, press Enter to skip): ')

# Also save Clerk Publishable Key for frontend
CLERK_PUBLISHABLE_KEY = getpass('Enter your Clerk PUBLISHABLE KEY (from Clerk Dashboard): ')

print("\n✓ All Clerk keys configured")
print(f"  - API Key: {'*' * 10}...")
print(f"  - JWKS URL: {'*' * 10}...")
print(f"  - Publishable Key: {'*' * 10}...")
print(f"  - Tavily API: {'*' * 10}..." if TAVILY_API_KEY else "  - Tavily API: (skipped)")
```

```python
import os

print("=" * 60)
print("CREATING ENVIRONMENT FILES")
print("=" * 60)

# Verify Clerk Publishable Key is set
if not CLERK_PUBLISHABLE_KEY or CLERK_PUBLISHABLE_KEY.strip() == '':
    print("\n⚠️  WARNING: CLERK_PUBLISHABLE_KEY is not set!")
    print("This will cause the frontend to show a configuration error.")
    CLERK_PUBLISHABLE_KEY = input('\nEnter your Clerk PUBLISHABLE KEY now: ')

# Backend .env file
env_content = f"""APP_NAME=Insurance Copilot
APP_VERSION=1.0.0
DEBUG=false
LOG_LEVEL=INFO
DATABASE_URL={DATABASE_URL}
DATABASE_URL_SYNC={DATABASE_URL_SYNC}
CLERK_API_KEY={CLERK_API_KEY}
CLERK_JWT_PUB_KEY={CLERK_JWT_PUB_KEY}
CLERK_WEBHOOK_SECRET=
TAVILY_API_KEY={TAVILY_API_KEY}
UPLOAD_DIR=uploads
MAX_UPLOAD_SIZE_MB=50
PGVECTOR_DIMENSION=384
CHUNK_SIZE=512
CHUNK_OVERLAP=64
TOP_K_RETRIEVAL=5
CORS_ORIGINS=["*"]
RATE_LIMIT_PER_MINUTE=60
RATE_LIMIT_PER_HOUR=1000
SENTRY_DSN="""

# Define the absolute path to ensure the file is created in the right place
env_path = '/content/AI-INSURANCE/BACKEND/.env'

# Ensure the directory exists first
os.makedirs(os.path.dirname(env_path), exist_ok=True)

with open(env_path, 'w') as f:
    f.write(env_content)

print("✓ Backend .env file created!")

# Frontend .env file - MUST include CLERK_PUBLISHABLE_KEY
if not CLERK_PUBLISHABLE_KEY or CLERK_PUBLISHABLE_KEY.strip() == '':
    print("\n❌ ERROR: CLERK_PUBLISHABLE_KEY is still empty!")
    print("The frontend will not work without this key.")
else:
    frontend_env = f"""VITE_CLERK_PUBLISHABLE_KEY={CLERK_PUBLISHABLE_KEY}
VITE_API_URL=http://localhost:8000
VITE_APP_NAME=Insurance Copilot"""

    frontend_env_path = '/content/AI-INSURANCE/FRONTEND/.env'
    os.makedirs(os.path.dirname(frontend_env_path), exist_ok=True)

    with open(frontend_env_path, 'w') as f:
        f.write(frontend_env)

    print("✓ Frontend .env file created!")

print("\n" + "=" * 60)
print("✓ Both backend and frontend environment files are configured!")
print("=" * 60)
```

## Step 9: Build Frontend and Start Backend

## Step 11: Setup Frontend Ngrok Tunnel and Start Dev Server

```python
from pyngrok import ngrok
import subprocess
import time

# Kill any existing ngrok tunnels
ngrok.kill()

# Create ngrok tunnel for frontend (port 5173 - Vite default)
frontend_tunnel = ngrok.connect(5173, bind_tls=True)
frontend_url = frontend_tunnel.public_url

print(f"\n{'='*60}")
print(f"FRONTEND PUBLIC URL: {frontend_url}")
print(f"BACKEND LOCAL URL: http://localhost:8000")
print(f"{'='*60}\n")
```

```python
import subprocess
import os
import time
from pyngrok import ngrok

os.chdir('/content/AI-INSURANCE/FRONTEND')

# Start the frontend dev server on port 5173
frontend_process = subprocess.Popen(
    ['npm', 'run', 'dev', '--', '--host', '0.0.0.0', '--port', '5173'],
    stdout=subprocess.PIPE,
    stderr=subprocess.STDOUT,
    text=True
)

time.sleep(5)

print(f"\n{'='*60}")
print(f"FRONTEND & BACKEND ARE RUNNING")
print(f"{'='*60}")
print(f"Frontend Public URL: {frontend_url}")
print(f"Frontend Local URL:  http://localhost:5173")
print(f"Backend URL:         http://localhost:8000")
print(f"API Docs:            http://localhost:8000/docs")
print(f"Health Check:        http://localhost:8000/health")
print(f"Database:            Local PostgreSQL")
print(f"LLM:                 Ollama ({OLLAMA_MODEL})")
print(f"{'='*60}\n")

try:
    for line in frontend_process.stdout:
        print(line, end='')
except KeyboardInterrupt:
    frontend_process.terminate()
    backend_process.terminate()
    ngrok.kill()
    print("\nFrontend and Backend stopped.")
```
